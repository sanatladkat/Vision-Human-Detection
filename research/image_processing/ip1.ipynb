```python
import cv2
import numpy as np
import os
import matplotlib.pyplot as plt
import ipywidgets as widgets
```

```python
BASE_DIR = os.getcwd()
FILE_NAME = r'garage5.png'
IMAGE_SAVE_PATH = os.path.join(BASE_DIR , 'images' , 'temp')
FILE_PATH = os.path.join(BASE_DIR,'images' , FILE_NAME)


print(IMAGE_SAVE_PATH)

```

```python
image = cv2.imread(FILE_PATH)
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
```

```python
hist = cv2.calcHist([image], [0], None, [256], [0, 256])
hist /= hist.sum()

plt.figure()
plt.title('Grayscale Histogram')
plt.xlabel('Pixel Value')
plt.ylabel('Frequency')
plt.plot(hist)
plt.xlim([0, 256])
plt.grid(axis='y')
plt.show()
```

```python


# Load the color image
image = cv2.imread(FILE_PATH)

# Split the image into color channels
b, g, r = cv2.split(image)

# Calculate the color histograms for each channel
hist_b = cv2.calcHist([b], [0], None, [256], [0, 256])
hist_g = cv2.calcHist([g], [0], None, [256], [0, 256])
hist_r = cv2.calcHist([r], [0], None, [256], [0, 256])

# Normalize the histograms
hist_b /= hist_b.sum()
hist_g /= hist_g.sum()
hist_r /= hist_r.sum()

# Create subplots
plt.figure(figsize=(12, 4))

# Blue channel histogram
plt.subplot(131)
plt.title('Blue Channel Histogram')
plt.xlabel('Pixel Value')
plt.ylabel('Frequency')
plt.plot(hist_b, color='blue')
plt.xlim([0, 256])
plt.grid(axis='y')

# Green channel histogram
plt.subplot(132)
plt.title('Green Channel Histogram')
plt.xlabel('Pixel Value')
plt.ylabel('Frequency')
plt.plot(hist_g, color='green')
plt.xlim([0, 256])
plt.grid(axis='y')

# Red channel histogram
plt.subplot(133)
plt.title('Red Channel Histogram')
plt.xlabel('Pixel Value')
plt.ylabel('Frequency')
plt.plot(hist_r, color='red')
plt.xlim([0, 256])
plt.grid(axis='y')

plt.tight_layout()  # Ensures that the subplots do not overlap
plt.show()
```

```python

image = cv2.imread(FILE_PATH)
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

plt.imshow(image)

```

```python
# Brightness


increased_brightness_image = cv2.convertScaleAbs(image, beta=100)
plt.imshow(increased_brightness_image)
```

```python
# Contrast Adjustment

alpha = 2.5  # Increase this value to increase contrast
contrast_adjusted_image = cv2.convertScaleAbs(image, alpha=alpha)
plt.imshow(contrast_adjusted_image)
```

```python
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
equalized_image = cv2.equalizeHist(gray_image)

plt.imshow(equalized_image)
```

```python
denoised_image = cv2.fastNlMeansDenoisingColored(image, None, h=10, hColor=1, templateWindowSize=7, searchWindowSize=21)

plt.imshow(denoised_image)
```

```python
gamma = 0.5  # Increase this value to make the image brighter
gamma_corrected_image = cv2.pow(image / 255.0, gamma)
gamma_corrected_image = (gamma_corrected_image * 255).astype('uint8')

plt.imshow(gamma_corrected_image)
```

```python
image = cv2.imread(FILE_PATH , 0)  # Load as grayscale

clahe = cv2.createCLAHE(clipLimit=2.0, tileGridSize=(8, 8))

# Apply CLAHE to the image
clahe_image = clahe.apply(image)

plt.imshow(cv2.cvtColor(clahe_image , cv2.COLOR_BGR2RGB))
```
